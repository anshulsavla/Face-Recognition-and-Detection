```python
face_cascade = cv2.CascadeClassifier('C:/Users/User-1/Desktop/Computer-Vision-Tutorial-master/Haarcascades/haarcascade_frontalface_default.xml')
def face_detector(img, size=0.5):
    #image = cv2.imread('C:\\Users\\User-1\\Desktop\\a\\DSC_0451.jpg')
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, 1.3, 5)


    if faces is ():
        return img
    
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w,y+h), (0,0,0), 20)
        #plt.imshow(image)
    return img



cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()
    cv2.imshow('test' , face_detector(frame))
    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break
        
cap.release()
cv2.destroyAllWindows()  
    
```
